Consider a network consisting of two neurons, given by
$$\begin{align}\tau\dot{u}_1(t) = -u_1(t) + \frac{s_1}{1+u_2(t)}, \\ 
   \tau \dot{u}_2(t) = -u_2(t) + \frac{s_2}{1 + u_1(t)}. \end{align}$$
The inputs $s_1$, $s_2$ are assumed to be non-negative.

Plot the system's dynamics for the two input signals using a quiver plot. Do the plots confirm your results from 1.2 and 1.3?

```python
### CODE FOR YOUR PLOTS HERE ###
import numpy as np
import matplotlib.pyplot as plt

# Define the system of equations
def system(u1, u2, s):
    tau = 1  # Assuming tau = 1 for simplicity
    du1 = -u1 + s / (1 + u2)
    du2 = -u2 + s / (1 + u1)
    return du1, du2

# Create a grid of points (u1, u2)
u1 = np.linspace(-1, 2, 20)
u2 = np.linspace(-1, 2, 20)
U1, U2 = np.meshgrid(u1, u2)

# Set up the figure and axes
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot for s = 0
s = 0
V1, V2 = system(U1, U2, s)
axes[0].quiver(U1, U2, V1, V2, color='r', alpha=0.6)
axes[0].set_title("Quiver Plot for s = 0")
axes[0].set_xlabel("u1")
axes[0].set_ylabel("u2")
axes[0].set_xlim(-1, 2)
axes[0].set_ylim(-1, 2)
axes[0].grid(False)

# Plot for s = 3/4
s = 3/4
V1, V2 = system(U1, U2, s)
axes[1].quiver(U1, U2, V1, V2, color='b', alpha=0.6)
axes[1].set_title("Quiver Plot for s = 3/4")
axes[1].set_xlabel("u1")
axes[1].set_ylabel("u2")
axes[1].set_xlim(-1, 2)
axes[1].set_ylim(-1, 2)
axes[1].grid(False)
plt.tight_layout()
plt.show()

### CODE FOR YOUR PLOTS HERE ###
```
